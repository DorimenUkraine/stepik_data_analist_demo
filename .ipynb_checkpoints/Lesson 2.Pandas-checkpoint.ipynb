Продолжим изучение данных и посмотрим на число поездок в выходные дни. Сгруппируйте данные по двум признакам: району города и является ли день выходным (колонки borough и hday). Сравните среднее число поездок, и выберите районы, из которых по праздникам в среднем поступает больше заказов, чем в обычные дни.

```python
taxi.groupby(['hday','borough']).agg({'pickups': 'mean'}).sort_values(by=['pickups'])
```

Для каждого района посчитайте число поездок по месяцам. Отсортируйте полученные значения по убыванию и сохраните результирующий датафрейм в pickups_by_mon_bor. Обратите внимание, что итоговый датасет должен состоять из 3-х колонок – pickup_month, borough, pickups. В задаче необходимо не забыть скинуть индексы (сделал в конце)

```python
pickups_by_mon_bor = taxi.groupby(['pickup_month','borough']).agg({'pickups': 'sum'}).sort_values(by=['pickups'], ascending=False).reset_index()
pickups_by_mon_bor
```

Напишите функцию temp_to_celcius, которая получает на вход колонку с температурой в °F, и возвращает значения, переведенные в градусы Цельсия.

```python
def temp_to_celcius(x):
    return ((x-32)*5)/9

taxi['temp_C'] = taxi['temp'].apply(temp_to_celcius)
```
